```python
from functions import *
np.set_printoptions(threshold=np.inf)

# Utiliza ordem alfabetica para descoberta dos ns
def BFS(inicial, final, matriz_adj):
    descobertos = []
    # Fila de vertices a serem explorados
    fila = []
    pais = dict()

    # Descobre o vertice inicial e o adiciona na fila para ser explorado
    descobertos.append(inicial)
    fila.insert(0, inicial)

    if inicial+1 > len(matriz_adj) or final+1 > len(matriz_adj):
        print("Vertice inicial ou final no existe no grafo")
        return []
    while True:
        # Obtem o ultimo item da fila
        v = fila.pop()
        # Obtm os vizinhos do vertice v
        vizinhos = [i for i in obter_vizinhos(v, matriz_adj) if i not in descobertos]
        for vizinho in vizinhos:
            # Descobre cada vizinho e os adiciona na fila para serem explorados
            descobertos.append(vizinho)
            fila.insert(0, vizinho)
            pais[vizinho] = v
            # Encerra o while se o vertice atual for igual ao destino
            if vizinho == final:
                return descobertos, pais

matriz_adj = gerar_matriz_adjacente("grafo.txt")
inicial, final = letter_to_int("G")-1, letter_to_int("S")-1 # Retorna o nmero correspondente a ordem alfabtica

descobertos, pais = BFS(inicial, final, matriz_adj)

imprimir_resultado(incrementar(descobertos), usarLetras=True)
```
