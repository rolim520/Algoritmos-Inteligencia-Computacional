```python
from functions import *
np.set_printoptions(threshold=np.inf)

# Utiliza ordem alfabetica para descoberta dos ns
def irrevogavel(inicial, final, matriz_adj):
    descobertos = []
    pais = dict()

    # Inicializa a variavel v
    v = inicial

    if inicial+1 > len(matriz_adj) or final+1 > len(matriz_adj):
        print("Vertice inicial ou final no existe no grafo")
        return []
    while True:
        descobertos.append(v)
        # Encerra o while se o vertice atual for igual ao destino
        if v == final:
            return descobertos, pais
        # Obtm a lista ordenada de vizinhos
        vizinhos = [i for i in obter_vizinhos(v, matriz_adj) if i not in descobertos]
        # Se no houver um vizinho no descoberto, o algoritmo encerra
        if vizinhos == []:
            print("Sem soluo!")
            return descobertos, pais
        pais[vizinhos[0]] = v
        # V se torna o vizinho de menor indice
        v = vizinhos[0]

matriz_adj = gerar_matriz_adjacente("grafo.txt")
inicial, final = letter_to_int("G")-1, letter_to_int("S")-1 # Retorna o nmero correspondente a ordem alfabtica -1

descobertos, pais = irrevogavel(inicial, final, matriz_adj)

imprimir_resultado(incrementar(descobertos), usarLetras=True)
```
