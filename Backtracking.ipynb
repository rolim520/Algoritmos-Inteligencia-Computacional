```python
from functions import *
np.set_printoptions(threshold=np.inf)

# Utiliza ordem alfabetica para descoberta dos ns
def backtracking(inicial, final, matriz_adj):
    descobertos = []
    # Pilha que guarda a ordem que os vertices devem ser descobertos
    pilha = []

    # Incializa a variavel v
    pilha.append(inicial-1)

    if inicial > len(matriz_adj) or final > len(matriz_adj):
        print("Vertice inicial ou final no existe no grafo")
        return []
    while True:
        # Retira o vertice no topo da pilha(Obs.: Pode conter vrtices repetidos)
        v = pilha.pop()
        if v not in descobertos:
            descobertos.append(v)
            # Encerra o while se o vertice atual for igual ao destino
            if v == final-1:
                return descobertos
            # Obtm a lista ordenada de vizinhos
            vizinhos = [i for i in obter_vizinhos(v, matriz_adj) if i not in descobertos]
            # Adiciona a lista de vizinhos em ordem inversa na pilha
            pilha += vizinhos[::-1]

matriz_adj = gerar_matriz_adjacente("grafo.txt")
inicial, final = letter_to_int("G"), letter_to_int("S") # Retorna o nmero correspondente a ordem alfabtica

descobertos = backtracking(inicial, final, matriz_adj)

imprimir_resultado(descobertos, usarLetras=True)
```
